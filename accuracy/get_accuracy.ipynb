```python
def get_cl_args():
    base_snap = "/global/cscratch1/sd/imit/gbruns/climate/TWEAKED_BEST_RUN/snapshot/TWEAKED_BEST_RUN/lr_1e-05_mu_0.9_step_20000_step-fac_0.99988478_clip_10000.0_decay_5e-05_rand_1_group_0_iter_20.caffemodel"
    #base_snap = "/project/projectdirs/dasrepo/gordon_bell/deep_learning/networks/climate/2d_semi_sup/models/vanilla__iter_3490.caffemodel"
    base_path = "/project/projectdirs/dasrepo/gordon_bell/deep_learning/networks/climate/2d_semi_sup/base_prototxt_files/deploy_vanilla_train.prototxt"
    cl_args = {"path_to_caffemodel":base_snap,
                "path_to_caffe_prototxt":base_path,
               "iterations": 19, "iou_thresh": 0.1}
    return cl_args

def get_accuracy(net, iterations, iou_thresh=0.5, reverse_conf=False):
    bba = BBox_Accuracy(iou_thresh=iou_thresh,reverse_conf=reverse_conf)
    for it in range(iterations):
        blobs =net.forward()
        bba = update_accuracy(bba, net,it)
        
        
    final_acc = bba.compute_final_accuracy()
    print "Final Accuracy for iou of %4.2f is %8.4f\n" % (bba.iou_thresh,final_acc)
    return final_acc
    


    

def update_accuracy(bba, net, it):
    label = np.float32(net.blobs["label"].data)
    net_output = get_net_output(net)
    bba.update_scores(net_output, label)
    print "Accuracy at iteration %i for iou thresh of %4.2f is %8.4f\n" % (it, bba.iou_thresh, bba.compute_final_accuracy())
    return bba
    



if __name__ == "__main__":

    os.environ["HDF5_DISABLE_VERSION_CHECK"] = "1"
    cl_args = get_cl_args()
    
    net = caffe.Net(cl_args["path_to_caffe_prototxt"],
                    cl_args["path_to_caffemodel"], 
                    caffe.TEST)
    
    
    get_accuracy(net, 
                 iterations=cl_args["iterations"],
                 iou_thresh=cl_args["iou_thresh"], reverse_conf=True)
    
    
    



    
```
